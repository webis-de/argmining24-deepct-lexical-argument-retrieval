```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
path = '/workspace/DeepCT_output/'
event = 'events.out.tfevents.1642032571.gammaweb04'
sum = 0
losses = []
for k, e in enumerate(tf.train.summary_iterator(path + event)):
    for v in e.summary.value:
        if (v.tag.startswith('loss_1')):
#             print(v)
            losses.append(v.simple_value)
            sum += 1

#         if (v.tag.endswith('super_puper_loss')):
#             losses.append(v.simple_value)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10, 5))
plt.plot(losses)
# plt.rcParams["figure.figsize"] = (30,5)
plt.xlabel('Every 1000th Step')
plt.ylabel('MSE Loss')
plt.legend(['Train'])
plt.show()
```
