```python
import tensorflow.compat.v1 as tf
import numpy as np
# tf.disable_v2_behavior()
```

```python
path = '/workspace/DeepCT_train_output/conclusions/'
event = 'events.out.tfevents.1642777916.gammaweb09'
sum = 0
losses = []
for k, e in enumerate(tf.train.summary_iterator(path + event)):
    for v in e.summary.value:
        if (v.tag.startswith('loss_1')):
            losses.append(v.simple_value)
            sum += 1
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10, 5))
plt.plot(losses)
# plt.rcParams["figure.figsize"] = (30,5)
plt.xlabel('Every 1000th Step')
plt.ylabel('MSE Loss')
plt.yticks(np.arange(0, 0.30, 0.02))
plt.xticks(np.arange(0, 1700, 100))
plt.legend(['Train'])
plt.title('DeepCT trained on Conclusions as Reference Field')
plt.show()
```
